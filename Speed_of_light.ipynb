```python
data['Time of rise for signal 2 (ns)'] = data['Time of rise for signal 2 (ns)']/10**(-9)
```

```python
data['Time of rise signal 1 (ns)'] = data['Time of rise signal 1 (ns)']/10**(-9)
```

```python
data['delta_t(ns)'] = (data['Time of rise for signal 2 (ns)'] - data['Time of rise signal 1 (ns)'])
```

```python
ftm = 0.3048
d1 = 1.625*ftm
data['delta_x(m)'] = (ftm*(data['Distance for cable 2 (ft)'])-d1).round(1)
```

```python
data.round(3)
```

Curve fitting from scatch

```python
N = 6
Delta = N*sum((data['delta_t(ns)'])**2) - (sum(data['delta_t(ns)']))**2
```

```python
A = (((sum((data['delta_t(ns)'])**2))*(sum(data['delta_x(m)'])))- ((sum((data['delta_t(ns)'])))*(sum((data['delta_t(ns)'])*(data['delta_x(m)'])))))/Delta
```

```python
B = ((N*(sum((data['delta_t(ns)'])*(data['delta_x(m)']))))-((sum(data['delta_x(m)']))*(sum(data['delta_t(ns)']))))/Delta
```

```python
Yi = A + B*data['delta_t(ns)']
```

```python
SttdY = np.sqrt((1/(N-2))*sum((data['delta_x(m)'] - A - B*data['delta_t(ns)'])**2))
SttdA = SttdY*np.sqrt((sum(data['delta_t(ns)'])**2)/Delta)
SttdB = SttdY*np.sqrt(N/Delta)
```

```python
data['dx_unc (m)']= (0.05*np.sqrt(2)).round(1)
data['dt_unc (ns)']= (0.1*np.sqrt(2)).round(1)
```
